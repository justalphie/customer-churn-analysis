## Exploratory data analysis
This part deals with the general information about the dataset: size, type of values, number of values, number of empty entries.

````python
import pandas as pd
df = pd.read_csv("data/BankChurners.csv")
````

````python
columns = df.columns.to_list()

for c in columns:
    print(c)
    print("Unique values count: ", df[c].nunique())
    print("Type of data: ", df[c].dtypes)
    print(df[c].unique())
    print("-------------------------")
````

For  ``` Education_Level, Marital_Status, Income_Category, Card_Category``` One-Hot encoding will be used, because they contain ```object``` values.
For ```Attrition_Flag, Gender``` columns binarization will be used. ```CLIENTNUM``` will be dropped.

````python
for c in columns:
    try:
        print(df[c].value_counts()["Unknown"], c)
    except:
        pass

````

Large proportion of data is missing. Dropping it will potentially lead to losing values. Therefore, these values will be encoded as a separate category.
