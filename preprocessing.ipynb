## Data preprocessing
After the exploratory data analysis, data is preprocessed: binarized and one-hot encoded.

```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
df = pd.read_csv("data/BankChurners.csv")
```

The categorical columns are encoded.

```python
object_columns = df.select_dtypes(include='object').columns
df = pd.get_dummies(df, columns=object_columns, dtype=float)
print(df.head())
```

Extra columns are dropped.

```python
df.drop(columns=['Attrition_Flag_Existing Customer', 'CLIENTNUM'], axis=1, inplace=True)
df.head()
```

The column to be predicted is inserted as the first one and renamed.

```python
col = df.pop('Attrition_Flag_Attrited Customer')
df.insert(0, col.name, col)
df.rename(columns={'Attrition_Flag_Attrited Customer': "Attrited"}, inplace=True)
```

Finally, the file is saved.

```python

df.to_csv("data/BankChurners_preprocessed.csv", index=False)
```
